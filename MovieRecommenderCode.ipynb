# Data Pre Processing

```python
import pandas as pd
import numpy as np
import ast 
```

```python
pd.reset_option('display.max_columns')
pd.reset_option('display.max_rows')
pd.reset_option('display.max_colwidth')
pd.reset_option('display.width')
movies= pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies= movies.merge(credits,left_on='title',right_on='title')
movies.head(1)
```

```python
## Columns to keep in the dataset 
#Genre
#ID
#Keywords
#Title
#Overview
#cast
#crew

movies=movies[['genres','id','keywords','title','overview','cast','crew']]
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]['genres']
```

```python
#To fetch only the name from Keywords and Genre
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(5)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(5)
```

```python
#To fetch only the cast 
def convert_cast(obj):
    L=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter !=3 :
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
#To fetch director
def fetch_Director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i["job"]=='Director'):
            L.append(i['name'])
    return L
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies['crew']=movies['crew'].apply(fetch_Director)
```

```python
movies.head(5)
```

```python
#To convert keywords from a string to a list so that we can concatenate it with other strings.
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
#To remove the space between words in each column.
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new=movies[['id','title','tags']]
```

```python
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new['tags'][0]
```

```python
new['tags']=new['tags'].apply(lambda x: x.lower())
new
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
## Function for Stemming the words in tags 

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))


    return " ".join(y)
        
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances= similarity[index]
    movies_list = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in movies_list[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Interstellar')
```
